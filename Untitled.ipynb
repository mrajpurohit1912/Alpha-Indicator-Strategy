```python
def calculate_rsi(data, column_name='intc', n=14):
    # Calculate daily price changes
    data['Price Change'] = data[column_name].diff()

    # Separate gains and losses
    gains = data['Price Change'].apply(lambda x: x if x > 0 else 0)
    losses = data['Price Change'].apply(lambda x: abs(x) if x < 0 else 0)

    # Calculate average gains and losses over n days
    avg_gain = gains.rolling(window=n, min_periods=1).mean()
    avg_loss = losses.rolling(window=n, min_periods=1).mean()

    # Calculate relative strength (RS)
    rs = avg_gain / avg_loss

    # Calculate RSI
    rsi = 100 - (100 / (1 + rs))

    # Add RSI column to the DataFrame
    data['RSI'] = rsi

    return data
```
